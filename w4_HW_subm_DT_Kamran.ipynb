```python
# -----------------------------------------------------------------
# Decision Tree Classifier
# Predict the income of an adult based on the census data
# -----------------------------------------------------------------

# Import libraries
import pandas as pd
import numpy as nm
```

```python
# Read dataset
df = pd.read_csv('decisiontreeAdultIncome.csv')
```

```python
df.head()
```

```python
# Check for Null values
pd.isnull(df).sum()
```

```python
# Create Dummy variables
df=pd.get_dummies(df)
```

```python
df.head()
```

```python
# Create X and Y Variables
X = df.drop(['IncomeClass_ <=50K','IncomeClass_ >50K'], axis=1)
X
```

```python
y = df['IncomeClass_ <=50K']
y
```

```python
# Split the X and Y dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Import and train classifier
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Test the model
predicted=model.predict(X_test)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
ac=accuracy_score(y_test,predicted)
ac
```

```python
matrix=confusion_matrix(y_test,predicted)
matrix
```
